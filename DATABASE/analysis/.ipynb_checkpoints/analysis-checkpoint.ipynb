```python
import time # this function allows me to record the length of time to run the code

start = time.time()

import pandas as pd 
import csv
import matplotlib as mat
import matplotlib.pyplot as plt
import psycopg2 as pg
import pandas.io.sql as psql
import numpy as np
import math

# setting input file as a variable to be easily used across the script

input_file = "mux.tsv" # input file name

# connect to our database and retrieve the necessary info

connection = pg.connect("dbname=d71uh4v1fd2hq user=tdsneouerzmxkj password=92a500cb091fe70168b32c66f\
a6a3d6c376d467d57fb9b663eb5d13446ecb2e6 host=ec2-54-75-245-94.eu-west-1.compute.amazonaws.com") # connects to database
dataframe = psql.read_sql('SELECT * FROM public."Phosphosite_table"', connection) # put database into pandas data frame
dataframe = dataframe[["GENE_NAME", "RESIDUE", "KINASE_NAME"]] # sets column headings 

# lists to extract the specific information that i want

names = []
residuelist = []
kinase = []
fclist = []

# basically cleaning the raw data to make it easier to use

with open(input_file, "r") as data: # reads input file varaible 
    data = csv.reader(data, delimiter='\t') # reads csv file 
    next(data, None) # skips header 
    for row in data: # loops through the files rows 
        data = row[0:6]
        if "(" in data[0]:
            name = data[0].split("(")[0] #separates the residue from name 
            residue1 = data[0].split("(")[1]
            residue = residue1.split(")")[0] # as above
            fc = data[3] # indexing the fold change 
        else:
            pass # prevents errors 
        if "_" in name: 
            name = name.split("_")[0] # removes "_HUMAN"
        
        for row in dataframe.itertuples():
            namez = row[1]
            resid = row[2]
            kin = row[3] # indexes important information
            if namez == name:
                if residue == resid:
                    names.append(name)
                    residuelist.append(resid)
                    kinase.append(kin)
                    fclist.append(fc) # appends all values to separate lists 
                    
# putting into pandas df

df = pd.DataFrame()
df["Names"] = names
df["Residue"] = residuelist
df["Kinase"] = kinase
df["Fold Change"] = fclist # turns these lists into a data frame 

# part two starts here

log2 = []
df = df[df != 'inf']
df = df[df != '0']
df = df[df != '']
df_cleaned = df.dropna(how='all')
df_cleaned2 = df_cleaned.dropna() # this cleans the dataframe further 

# again, cleaning the data
# calcualting the log2 for the substrates

for row in df_cleaned2.itertuples():
#     name = row[1]
#     residue = row [2]
#     kinase = row[3]
    fc = float(row[4]) # floats the fc value 
    twoval = math.log(fc, 2) # finds the log2 of the fc value 
    log2.append(twoval) # appends log2(fc) into a alist 
    
# creating a new column containing the log2(fc)

df_cleaned2["log2(fc)"] = log2 # list is integrated into data frame 

# calculating log2(fc) specific to kinases

total = {}

for row in df_cleaned2.itertuples():
    name = row[1]
    residue = row [2]
    kinase = row[3]
    fc = float(row[4])
    log2fc = float(row[5])
    if kinase == kinase:
        if kinase in total:
            total[kinase].append(log2fc) # appends to already existing key 
        else:
            total[kinase] = [log2fc] # creates a new key with first value 

# taking the mean value

newdict = {}

for key, value in total.iteritems():
    s = sum(value)/len(value) # this finds the mean of the values per key
    newdict[key] = s # appends these to a new dictionary 
    
# reading in raw data into pandas df & cleaned again 

rdf = pd.read_csv(input_file, sep="\t")    # puts raw data file in a data frame   
nrdf = rdf.iloc[:,0:4]
new = nrdf.replace(0, np.nan)
new = new.dropna(how='all')
new = new.dropna() # this cleans the data 

# calcuating total log2(fc) mean value for all phosphosites

another = []

for d in new.itertuples():
    fc = str(d[4]) # converted to string for next condition
    if fc != 'inf': # removes uninformative data
        fc = math.log(float(fc), 2) # finds the log2 of all fc 
        another.append(fc) # appends it to a list 
        
p = sum(another)/len(another) # finds the mean of the total log2(fc) values 

# total number of phosphosites specific to kinases, square rooted

m = math.sqrt(len(df_cleaned2.index)) # this finds the square root of number of substrates in our raw data 

# mean log2(fc) of all phosphosites 

s = np.std(another) # the standard deviation of all log2(Fc) in our raw data 

# allocating the z-scores to their kinase

score = {}

for key, value in newdict.iteritems():
    val = ((value - p)*m)/float(s) # this calculates the z-score
    score[key] = val # appends it to a new dictionary 

# putting values into a pandas df    
    
final = pd.DataFrame(score.items()) # creates a new data frame 
final.columns = ["Kinase", "z-score"] # sets data frame column names 

# putting kinases in order from lowest to highest z-score

organised = final.sort_values(['z-score'], ascending = True) 

# indexing top ten z-scores

top10 = organised.iloc[0:11,:] 

# putting values into a bar graph

ax = top10.plot.bar(x = "Kinase", rot=0, figsize=(7,3.5)) 

# adjusting figure size

plt.subplots_adjust(bottom=0.35, left=0.35)

# setting title

ax.set_title('Top Ten Downregulated Kinases')

# removing figure legend

ax.get_legend().remove()

# setting y-label

ax.set_ylabel('z-score')

# adjusting x-axis labels to be more visible

plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')

# converts into figure

fig = ax.get_figure()

# saves figure

fig.savefig("topten.png")

# saves data into csv files

organised.to_csv('total_list.csv')
top10.to_csv('topten_list.csv')

end = time.time()
print str((end - start)/60) + " Minutes"
```
