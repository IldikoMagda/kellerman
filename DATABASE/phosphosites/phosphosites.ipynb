```python
import csv
import re
from pyensembl import EnsemblRelease
import urllib
from IPython.lib.pretty import pprint
import pandas as pd
import pprint
import os
import time
```

```python
accession = []
with open("uniprot-reviewed_yes+taxonomy_9606.csv", "r") as csv_file: # opens file with IDs and names 
    csv_reader = csv.reader(csv_file) # reads the file 
    next(csv_reader) # skips the header 
    for row in csv_reader:
        row[1] = row[1].split("_")[0] # removes "_HUMAN"
        accession.append(row[0] + "," + row[1]) # adds to list 
```
