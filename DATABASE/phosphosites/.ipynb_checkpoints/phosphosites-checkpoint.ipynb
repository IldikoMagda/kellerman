```python
import csv
import re
from pyensembl import EnsemblRelease
import urllib
from IPython.lib.pretty import pprint
import pandas as pd
import pprint
import os
```

```python
newlist =[]
protein =[]
cleaned =[]
noHUMAN =[]

with open('az20.tsv', 'rb') as csvfile:
    read = csv.reader(csvfile, delimiter='\t')
    for row in read:
        newlist.append(row[0])
    better = newlist[1:]

for g in better:
    h = g.split("(")
    f = h[0]
    if "_HUMAN" in f:
        protein.append(f)
    else:
        new = f + "_HUMAN"
        protein.append(new)
 
for n in protein:
    if n not in cleaned:
        cleaned.append(n)
               
protein_file = open('protein_list_new.csv', 'w')
proteinlist = open ("protein-only.csv", "w")
for l in cleaned:
    line = l + "\n"
    protein_file.write(line)
    so = l.split("_")[0]
    noHUMAN.append(so)

for t in noHUMAN:
    lines = t + "\n"
    proteinlist.writelines(lines)
    
proteinlist.close()
protein_file.close()
```

```python
import urllib
import re 

mykinase = []
myresidue = []
counter = -1
myfile = open('kinase_list_mod.csv', 'w')
header = ' ' + ',' + 'Target' + ',' + 'Position' + ',' + 'Residue' + ',' + 'Kinase' +'\n'
myfile.write(header)
for y in cleaned:
    print counter
    h = str(y)
    URL = 'https://www.uniprot.org/uniprot/?query=' + h +'&columns=feature(MODIFIED_RESIDUE)&format=tab'
    df = urllib.urlopen(URL).read()
    dfg = str(df)
    we = dfg.split('MOD_RES')
    counter = counter + 1
    for x in we:
        if 'Phospho' in x:
            if 'by' in x:
                trap = []
                aa = x.split('by')
                pre_kinase = aa[1].split('.')[0]
                gg = re.findall(r"[\w']+", pre_kinase)
                for t in gg:
                    if t != "and":
                        if t != "alternate":
                            if t != "in":
                                if t != "vitro":
                                    if t != "or":
                                        if t != "autocatalysis":
                                            if t != "host":
                                                kinase = aa[1].split('.')[0]
                                                position = aa[0].split(' ')[1]
                                                pre_pre_residue = aa[0].split(' ')[3]
                                                pre_residue = pre_pre_residue[7:]
                                                residue = pre_residue[0:3].upper()
                                                line = str(counter) + ',' + str(y) + ',' + str(position) + ',' + str(residue) + ',' + str(t) + '\n'
                                                myfile.write(line)
                
myfile.close()
```
