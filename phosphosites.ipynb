```python
import csv
import re
from pyensembl import EnsemblRelease
import urllib
from IPython.lib.pretty import pprint
```

```python
newlist =[]
protein =[]
cleaned =[]
noHUMAN =[]

with open('az20.tsv', 'rb') as csvfile:
    read = csv.reader(csvfile, delimiter='\t')
    for row in read:
        newlist.append(row[0])
    better = newlist[1:]

for g in better:
    h = g.split("(")
    f = h[0]
    if "_HUMAN" in f:
        protein.append(f)
    else:
        new = f + "_HUMAN"
        protein.append(new)
 
for n in protein:
    if n not in cleaned:
        cleaned.append(n)
               
protein_file = open('protein_list_new.csv', 'w')
proteinlist = open ("protein-only.csv", "w")
for l in cleaned:
    line = l + "\n"
    protein_file.write(line)
    so = l.split("_")[0]
    noHUMAN.append(so)

for t in noHUMAN:
    lines = t + "\n"
    proteinlist.writelines(lines)
    
proteinlist.close()
protein_file.close()
```
